```python
train = pd.read_csv('input/train.csv')
test = pd.read_csv('input/test.csv')
```

```python
train = train.drop(columns = 'Name')
test = test.drop(columns = 'Name')
```

```python
sex_mapping = {"male":0, "female":1}

train['Sex'] = train['Sex'].map(sex_mapping)
test['Sex'] = test['Sex'].map(sex_mapping)
```

```python
train['Age'] = train['Age']/80
test['Age'] = test['Age']/80
```

```python
train = train.drop(columns = ['Embarked', 'Cabin', 'Ticket'])
test = test.drop(columns = ['Embarked', 'Cabin', 'Ticket'])
```

```python
train['Fare'] = train['Fare']/513
test['Fare'] = test['Fare']/513
```

```python
train = train.drop(columns = ['PassengerId'])
```

```python
train.loc[train['Parch'] == 0, 'Parch'] = 2
train.loc[train['Parch'] == 1, 'Parch'] = 0
train.loc[train['Parch'] == 2, 'Parch'] = 0
train.loc[(train['Parch'] > 2) & (train['Parch'] < 4), 'Parch'] = 1
train.loc[train['Parch'] >= 4, 'Parch'] = 3

test.loc[test['Parch'] == 0, 'Parch'] = 2
test.loc[test['Parch'] == 1, 'Parch'] = 0
test.loc[test['Parch'] == 2, 'Parch'] = 0
test.loc[(test['Parch'] > 2) & (test['Parch'] < 4), 'Parch'] = 1
test.loc[test['Parch'] >= 4, 'Parch'] = 3
```

```python
train.loc[train['SibSp'] == 0, 'SibSp'] = 2
train.loc[train['SibSp'] == 1, 'SibSp'] = 0
train.loc[train['SibSp'] == 2, 'SibSp'] = 0
train.loc[(train['SibSp'] > 2) & (train['SibSp'] < 4), 'SibSp'] = 1
train.loc[train['SibSp'] >= 4, 'SibSp'] = 3

test.loc[test['SibSp'] == 0, 'SibSp'] = 2
test.loc[test['SibSp'] == 1, 'SibSp'] = 0
test.loc[test['SibSp'] == 2, 'SibSp'] = 0
test.loc[(test['SibSp'] > 2) & (test['SibSp'] < 4), 'SibSp'] = 1
test.loc[test['SibSp'] >= 4, 'SibSp'] = 3
```

```python
train_with_age = train.loc[train['Age'].isnull() == False]
```

```python
test_with_age = test.loc[test['Age'].isnull() == False]

test_with_fare = test.loc[test['Fare'].isnull() == False]

test_with_age_fare = test_with_age.loc[test_with_age['Fare'].isnull() == False]
```

```python
train_no_age = train.drop(columns = 'Age')
train_no_fare = train_with_age.drop(columns = 'Fare')
train_no_age_fare = train.drop(columns = ['Age', 'Fare'])

test_no_age = test.loc[test['Age'].isnull()]
test_no_fare = test.loc[test['Fare'].isnull()]

test_no_age = test_no_age.drop(columns = 'Age')
test_no_fare = test_no_fare.drop(columns = 'Fare')
test_no_age_fare = test.drop(columns = ['Age', 'Fare'])
```

```python
train_with_age_data = train_with_age.drop(columns = 'Survived')
train_with_age_target = train_with_age['Survived']

train_no_age_data = train_no_age.drop(columns = 'Survived')
train_no_age_target = train_no_age['Survived']

train_no_fare_data = train_no_fare.drop(columns = 'Survived')
train_no_fare_target = train_no_fare['Survived']

train_no_age_fare_data = train_no_age_fare.drop(columns = 'Survived')
train_no_age_fare_target = train_no_age_fare['Survived']
```

```python
k_fold = KFold(n_splits = 10, shuffle = True, random_state = 0)
```

```python
clf = KNC(n_neighbors = 4)
scoring = 'accuracy'
score = cross(clf, train_with_age_data, train_with_age_target, cv = k_fold, n_jobs = 1, scoring = scoring)
print("with age and fare",score)
print(round(np.mean(score)*100, 2))

clf = KNC(n_neighbors = 4)
scoring = 'accuracy'
score = cross(clf, train_no_age_data, train_no_age_target, cv = k_fold, n_jobs = 1, scoring = scoring)
print("no age" , score)
print(round(np.mean(score)*100, 2))

clf = KNC(n_neighbors = 4)
scoring = 'accuracy'
score = cross(clf, train_no_fare_data, train_no_fare_target, cv = k_fold, n_jobs = 1, scoring = scoring)
print("no fare",score)
print(round(np.mean(score)*100, 2))

clf = KNC(n_neighbors = 4)
scoring = 'accuracy'
score = cross(clf, train_no_age_fare_data, train_no_age_fare_target, cv = k_fold, n_jobs = 1, scoring = scoring)
print("no fare",score)
print(round(np.mean(score)*100, 2))
```

```python
clf = KNC(n_neighbors = 5)
scoring = 'accuracy'
score = cross(clf, train_with_age_data, train_with_age_target, cv = k_fold, n_jobs = 1, scoring = scoring)
print("with age and fare",score)
print(round(np.mean(score)*100, 2))

clf = KNC(n_neighbors = 5)
scoring = 'accuracy'
score = cross(clf, train_no_age_data, train_no_age_target, cv = k_fold, n_jobs = 1, scoring = scoring)
print("no age" , score)
print(round(np.mean(score)*100, 2))

clf = KNC(n_neighbors = 5)
scoring = 'accuracy'
score = cross(clf, train_no_fare_data, train_no_fare_target, cv = k_fold, n_jobs = 1, scoring = scoring)
print("no fare",score)
print(round(np.mean(score)*100, 2))

clf = KNC(n_neighbors = 5)
scoring = 'accuracy'
score = cross(clf, train_no_age_fare_data, train_no_age_fare_target, cv = k_fold, n_jobs = 1, scoring = scoring)
print("no fare",score)
print(round(np.mean(score)*100, 2))
```

```python
for i in range(1,20):

    clf = KNC(n_neighbors = i)
    scoring = 'accuracy'
    score1 = cross(clf, train_with_age_data, train_with_age_target, cv = k_fold, n_jobs = 1, scoring = scoring)
    # print("with age and fare",score)
    # print(round(np.mean(score)*100, 2))

    clf = KNC(n_neighbors = i)
    scoring = 'accuracy'
    score2 = cross(clf, train_no_age_data, train_no_age_target, cv = k_fold, n_jobs = 1, scoring = scoring)
    # print("no age" , score)
    # print(round(np.mean(score)*100, 2))

    clf = KNC(n_neighbors = i)
    scoring = 'accuracy'
    score3 = cross(clf, train_no_fare_data, train_no_fare_target, cv = k_fold, n_jobs = 1, scoring = scoring)
    # print("no fare",score)
    # print(round(np.mean(score)*100, 2))

    clf = KNC(n_neighbors = i)
    scoring = 'accuracy'
    score4 = cross(clf, train_no_age_fare_data, train_no_age_fare_target, cv = k_fold, n_jobs = 1, scoring = scoring)
    # print("no fare",score)
    print("for i =",i,"\n wAF: ", round(np.mean(score1)*100, 2), ", nA: ", round(np.mean(score2)*100, 2),\
         ", nF: ",round(np.mean(score3)*100, 2), ", nAF: ", round(np.mean(score3)*100, 2),\
         ", Overall: ", sum([round(np.mean(score4)*100, 2),round(np.mean(score1)*100, 2),round(np.mean(score2)*100, 2),round(np.mean(score3)*100, 2)]))
```
